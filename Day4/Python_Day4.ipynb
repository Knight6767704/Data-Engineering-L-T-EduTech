```python
#dfs
visited=set()
def dfs(visisted,graph,node):
  if node not in visited:
    print(node)
    visisted.add(node)
    for neighbour in graph[node]:
      dfs(visited,graph,neighbour)
graph = {'0':['1','2'],'1':['3','4'],'2':['5','6'],'3':[],'4':[],'5':[],'6':[]}
dfs(visited,graph,'0')
```

```python
#count number of sink nodes in this graph
graph = {'0':['1','2'],'1':['3','4'],'2':['5','6'],'3':[],'4':[],'5':[],'6':[]}
count=0
for node in graph:
  print(node,graph[node])
  if graph[node]:
    for n in graph[node]:
      print(n)
  else:
    count+=1
print("----------")
print(count)
print("----------")
```
